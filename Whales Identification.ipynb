## Download the Pretrained Model

```python
from tensorflow import keras
```

```python
base_model1 = keras.applications.VGG16(
    weights='imagenet',  # load weight pre-trained on ImageNet
    input_shape=(224, 224, 3),
    include_top=False  # remove the last later of the pre-trained model, so we can add new layers
)
```

```python
base_model2 = keras.applications.VGG19(
    weights='imagenet',  # load weight pre-trained on ImageNet
    input_shape=(224, 224, 3),
    include_top=False  # remove the last later of the pre-trained model, so we can add new layers
)
```

```python
base_model3 = keras.applications.ResNet50(
    weights='imagenet',  # load weight pre-trained on ImageNet
    input_shape=(224, 224, 3),
    include_top=False  # remove the last later of the pre-trained model, so we can add new layers
)
```

```python
base_model4 = keras.applications.ResNet101(
    weights='imagenet',  # load weight pre-trained on ImageNet
    input_shape=(224, 224, 3),
    include_top=False  # remove the last later of the pre-trained model, so we can add new layers
)
```

## Adding New Layers

```python
inputs=keras.Input(shape=(224,224,3))
#separately from setting trainable on the model, we set training to False
x1=base_model1(inputs,training=False)
x1=keras.layers.GlobalAveragePooling2D()(x1)
#including new_whale ID, there are 5005 unique categories (whale_IDs)
#A Dense classifier with 5004 units 
outputs1=keras.layers.Dense(5005)(x1)
model1=keras.Model(inputs,outputs1)
```

```python
inputs=keras.Input(shape=(224,224,3))
#separately from setting trainable on the model, we set training to False
x1=base_model1(inputs,training=False)
x1=keras.layers.GlobalAveragePooling2D()(x1)
#in subset dataset, there are 273 unique categories (whale_IDs)
#A Dense classifier with 5004 units 
outputs1_subset=keras.layers.Dense(273)(x1)
model1_subset=keras.Model(inputs,outputs1_subset)
```

```python
inputs=keras.Input(shape=(224,224,3))
#separately from setting trainable on the model, we set training to False
x4=base_model4(inputs,training=False)
x4=keras.layers.GlobalAveragePooling2D()(x4)
#including new_whale ID, there are 5005 unique categories (whale_IDs)
#A Dense classifier with 5004 units 
outputs4=keras.layers.Dense(5005)(x4)
model4=keras.Model(inputs,outputs4)
```

Try model4 (ResNet101) on grayscale data.

```python
from tensorflow.keras.layers import Concatenate

inputs=keras.Input(shape=(224,224,1))
img_conc = Concatenate()([inputs, inputs, inputs]) # repeat the greyscale color channle 3 times to mimic RGB in model
#separately from setting trainable on the model, we set training to False

base_model4_gray = keras.applications.ResNet101(
    weights='imagenet',  # load weight pre-trained on ImageNet
    input_shape=(224, 224, 3),
    include_top=False,  # remove the last later of the pre-trained model, so we can add new layers
    input_tensor=img_conc
)

x4=base_model4_gray(inputs,training=False)
x4=keras.layers.GlobalAveragePooling2D()(x4)
#including new_whale ID, there are 5005 unique categories (whale_IDs)

base_model4_gray.trainable=False #ResNet101

#A Dense classifier with 5004 units 
outputs4=keras.layers.Dense(5005)(x4)
model4=keras.Model(inputs,outputs4)
```

Try Max pooling instead of Average Pooling while adding layers to ResNet50.

```python
inputs=keras.Input(shape=(224,224,3))
#separately from setting trainable on the model, we set training to False
x3=base_model3(inputs,training=False)
x3=keras.layers.GlobalMaxPooling2D()(x3)
#including new_whale ID, there are 5005 unique categories (whale_IDs)
#A Dense classifier with 5004 units 
outputs3=keras.layers.Dense(5005)(x3)
model3=keras.Model(inputs,outputs3)
```

## Examing the Predictions

```python
from tensorflow.keras.preprocessing import image as image_utils
from tensorflow.keras.applications.imagenet_utils import preprocess_input
```

Prediction function for modeling using only existing whales.

```python
predictions = []
paths = train_map_df['Image']
existing_ids = train_map_df_existing['Id'].unique()

def make_prediction(image_path):
    image = image_utils.load_img(os.path.join('data', 'train', image_path), target_size=(224, 224), color_mode="rgb")
    image = image_utils.img_to_array(image)
    image = image.reshape(1,224,224,3)
    image = preprocess_input(image)
    preds = model4_existing.predict(image)
    predictions.append(existing_ids[np.argsort(preds)[-5:][::-1]])
    return preds[0][0]
```

```python
for id in train_map_df_existing['Image']:
    predictions.append(make_prediction(id))
```
